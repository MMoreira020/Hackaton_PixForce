```python
import os
import json
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Defina o diretório alvo
target_dir = r"C:\Users\maste\Downloads\hackathon\hackathon\futebol\2023_06_19_Mirassol_ABC\targets"

# Inicialize listas para armazenar rótulos e previsões
true_labels = []  # rótulos verdadeiros de cada frame
predictions = []  # previsões com base na lógica definida

# Função para definir o critério de classificação
def classify_frame(data):
    # Exemplo: se houver mais de 5 pessoas detectadas, classifique como 'multidão'
    num_people = len(data['2dPlayerTracking']['people'])
    if num_people > 5:
        return 'multidão'
    else:
        return 'pequeno_grupo'

# Loop para ler e classificar cada arquivo JSON
for filename in os.listdir(target_dir):
    if filename.endswith('.json'):
        with open(os.path.join(target_dir, filename), 'r') as file:
            data = json.load(file)
            prediction = classify_frame(data)
            predictions.append(prediction)
            # true_labels.append(label)  # Substitua por rótulos reais se disponíveis

# Cálculo das métricas
accuracy = accuracy_score(true_labels, predictions)
precision = precision_score(true_labels, predictions, average='weighted')
recall = recall_score(true_labels, predictions, average='weighted')
f1 = f1_score(true_labels, predictions, average='weighted')

print("Acurácia:", accuracy)
print("Precisão:", precision)
print("Revocação:", recall)
print("F1 Score:", f1)
```
